```python
import pandas as pd
data = pd.read_csv('marriage.csv')
print(data.shape)
data.head()
```

```python
data.isnull().sum()
```

```python
(data.shape[0] - data.dropna().shape[0])/data.shape[0]
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.head(2)
```

```python
X = data.loc[:, ['gender', 'height', 'religion', 'caste', 'mother_tongue', 'country']]
y = data.age_of_marriage
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
X.loc[:, ['gender', 'religion', 'caste', 'mother_tongue', 'country']] = X.loc[:, ['gender', 'religion', 'caste', 'mother_tongue', 'country']].apply(enc.fit_transform)
```

```python
X.head()
```

```python
int(X.loc[1,'height'].split('\'')[0])*30.48
```

```python
int(X.loc[1,'height'].split('\'')[1].replace('"', ''))*2.54
```

```python
def h_cms(h):
  return int(h.split('\'')[0])*30.48 + int(h.split('\'')[1].replace('"', ''))*2.54
```

```python
X['height_cms'] = X.height.apply(h_cms)
```

```python
X.head()
```

```python
X.drop('height', inplace=True, axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=80, max_depth=11)
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

# Evaluation

```python
from sklearn.metrics import mean_absolute_error, r2_score
print("MAE :", mean_absolute_error(y_test, y_predict))
r2_score(y_test, y_predict)
```

# Export model

```python
import pickle
# Saving model
pickle.dump(model, open('model.pkl', 'wb'))
```
